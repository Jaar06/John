### Print numbers with a 3 in Them

```python
for x in range(1,1001): 
  hasThree = False 
  holder = x 

  #check if it has a three 
  while(holder > 0): 
    if(holder%10 == 3):
      hasThree= True 
      break 
    holder = holder // 10

  if(hasThree): 
    print(x,end=" ")
```

### Show numbers whose digits add to 17

```python
for x in range(1,100001): 
  total = 0 
  holder = x 

  #add the digits together 
  while(holder >0): 
    total += holder%10 
    holder = holder//10

  if(total == 17):
    print(x, end=" ")
```

### Show numbers whose digits add to 23 and have a 3 in them

```python
#pull the digits from a number 
for i in range(1,10001):
  total = 0
  hasThree = False 
  x = holder = i

  #check if it has a three 
  while(holder > 0): 
    if(holder%10 == 3):
      hasThree= True 
      break 
    holder = holder // 10

  #add the digits together
  while(hasThree and x >0): 
    total += x%10
    x = x//10

  if(total == 23):
    print(i, end=" ")
```
